```python
import pandas as pd
```

```python
dataRaw = pd.read_csv('data/HousingPrice/HousingPriceOriginal/house-pricing-2022.csv')
```

```python
#Data Preprocessing
#Data Cleaning
#Data Profiling
dataProfiled = dataRaw[['389995','CV22 5SH','RUGBY','RUGBY.1','WARWICKSHIRE']]
dataProfiled.columns = ['Price', 'Post Code', 'City', 'District','County']
```

```python
filtered_data = dataProfiled[dataProfiled['County'].str.contains('Yorkshire|Oxfordshire', case=False)]
filtered_data.head(100)
filtered_data.to_csv('data/HousingPrice/HousingPriceCleaned/profiled_data.csv', index=False)
```
