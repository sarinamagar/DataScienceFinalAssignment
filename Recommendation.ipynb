```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
house_price = pd.read_csv('data/HousingPrice/HousingPriceCleaned/profiled_data.csv')
population_data = pd.read_csv('data/Population/Cleaned_Town_population.csv')
LSOA = pd.read_csv('data/LSOA/Postcode_to_LSOA.csv',encoding='latin')
school_data = pd.read_csv('data/School/school_data_2021-2022_cleaned.csv')
broadband_performance_data = pd.read_csv('data/Broadband/BroadbandCleaned/broadband_filtered_performance_all.csv')
broadband_coverage_date = pd.read_csv('data/Broadband/BroadbandCleaned/broadband_filtered_coverage_all.csv')
crime_data = pd.read_csv('data/crime/crime_merged_data.csv')
```

```python
recommendation = pd.DataFrame()
LSOA.head()
```

```python

prices = house_price['Price'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1000))
price_scores = scaler.fit_transform(prices)
reversed_scores = 10 - price_scores
house_price['price_score'] = reversed_scores
house_price.head(10)
```

```python
population = population_data['Population2022'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 10))
score = scaler.fit_transform(population)
reversed_scores = 10 - score
population_data['population_score'] = reversed_scores
population_data.rename(columns={'PostCode': 'Post Code'}, inplace=True)
population_data = population_data.drop(columns=['Year'])
population_data.head()
```

```python
attainment = school_data['Attained'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 10))
score = scaler.fit_transform(attainment)
school_data['school_score'] = score
school_data.head()
```

```python
performance = broadband_performance_data['Average download speed (Mbit/s)'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 10))
score = scaler.fit_transform(performance)
broadband_performance_data['performance_score'] = score
broadband_performance_data.rename(columns={'postcode_space': 'Post Code'}, inplace=True)
broadband_performance_data.head()
```

```python
coverage = broadband_coverage_date['UFBB availability (% premises)'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 10))
score = scaler.fit_transform(coverage)
broadband_coverage_date['coverage_score'] = score
broadband_coverage_date.rename(columns={'postcode_space': 'Post Code'}, inplace=True)
post_code_counts = broadband_coverage_date.groupby('Post Code').size()
broadband_coverage_date.head()
```

```python
data = pd.read_csv('data/crime/crime_merged_data_cleaned.csv')


postcode = pd.read_csv('data/LSOA/Postcode_to_LSOA.csv',encoding='latin')

housing_data = pd.read_csv('data/HousingPrice/HousingPriceCleaned/profiled_data.csv')

fist_merge = postcode.merge(housing_data,left_on='pcd8',right_on='Post Code', how='inner')
fist_merge = fist_merge[['Post Code', 'City', 'District', 'County','lsoa11cd']]
fist_merge.rename(columns={'lsoa11cd': 'LSOA'}, inplace=True)
final_merge = fist_merge.merge(data,left_on='LSOA', right_on='LSOA code', how='inner' )
final_merge.drop(columns={'LSOA'}, inplace=True)
crime_counts = final_merge.groupby(['Post Code', 'Crime type'])['Crime type'].count().reset_index(name='Total Occurrences by Crime Type')
total_occurrences_by_postal_code = final_merge.groupby('Post Code')['Crime type'].count().reset_index(name='Total Occurrences by Postal Code')
crime_counts = crime_counts.merge(total_occurrences_by_postal_code, on='Post Code', how='left')
unique_postcodes = total_occurrences_by_postal_code.drop_duplicates(subset='Post Code')
unique_postcodes = unique_postcodes[['Post Code', 'Total Occurrences by Postal Code']]
unique_postcodes.head(5)


```

```python
crime = unique_postcodes['Total Occurrences by Postal Code'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 100))
score = scaler.fit_transform(crime)
reversed_scores = 10 - score
unique_postcodes['crime_score'] = reversed_scores
unique_postcodes.head()
```

```python
recommendation['Post Code'] = house_price['Post Code']
recommendation['Price Score'] = house_price['price_score']
recommendation['Population Score'] = population_data['population_score']
recommendation['School Score'] = school_data['school_score']
recommendation['Performance Score'] = broadband_performance_data['performance_score']
recommendation['Coverage Score'] = broadband_coverage_date['coverage_score']
recommendation['Crime Score'] = unique_postcodes['crime_score']
recommendation.dropna(inplace=True)
recommendation.head(20000)
```

```python
recommendation['Final Score'] = (
    recommendation['Price Score'] +
    recommendation['Population Score'] +
    recommendation['School Score'] +
    recommendation['Performance Score'] +
    recommendation['Coverage Score'] +
    recommendation['Crime Score']
)
recommendation.head(10)
```
